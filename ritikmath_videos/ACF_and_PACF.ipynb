```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
register_matplotlib_converters()
```

Ice-cream data

```python
df_ice_cream=pd.read_csv('ice_cream.csv')
```

```python
df_ice_cream.head()
```

```python
df_ice_cream.rename(columns={'DATE':'date','IPN31152N':'production'},inplace=True)
```

```python
df_ice_cream['date']=pd.to_datetime(df_ice_cream.date)
```

```python
#setting date column as datetime
df_ice_cream.set_index('date',inplace=True)
```

```python
# taking dataset from 2010 onwards only
satat_date=pd.to_datetime('2010-01-01')
df_ice_cream=df_ice_cream[satat_date:]
```

```python
df_ice_cream.head()
```

```python
# Plotting the dataset 
plt.figure(figsize=(10,4))
plt.plot(df_ice_cream.production)
plt.title('Ice Cream Production [2010-2020]')
plt.ylabel('Production')
plt.xlabel('Time')
for year in range(2010,2021):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'),color='k',linestyle='--',alpha=0.2)
```

```python
acf_plot=plot_acf(df_ice_cream.production,lags=100)
#An int or array of lag values, used on horizontal axis. Uses np.arange(lags) when lags is an int. If not provided, lags=np.arange(len(corr)) is used.
```

0th index acf and pcf have acf and pacf same =1 (ofc because it is correlation with itself)

```python
pacf_plot=plot_pacf(df_ice_cream.production)
```
